```python
from tensorflow.keras.datasets import mnist
(x_coordinates_train, y_coordinates_train), (x_coordinates_test, y_coordinates_test) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np

x ,a = plt.subplots(10, 5)
def check(y_coordinates_train, y, i):
  if i==y_coordinates_train[y]:
    return True

for classnum in range(10):
    classet = [ind for ind in range(60000) if check(y_coordinates_train, ind, classnum)]
    count = 0
    # print(classet)
    for mnind  in  (rd.sample(classet, 5)):
        a[classnum, count].imshow(x_coordinates_train[mnind], cmap='gray')
        count = count+1
plt.show()
```
